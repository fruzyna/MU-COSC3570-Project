```python
import pandas as pd
```

```python
crimes = pd.read_csv('data/crimes.csv', ',', parse_dates = True, index_col='ID')
crimes.head()
```

```python
crimeplaces = crimes['Location Description'].value_counts()
ax = crimeplaces[:25].plot(kind='bar', figsize=(10, 5))
ax.set_xlabel('Location Description')
ax.set_ylabel('Instances')
ax.set_title('Top 25 Most Common Location Descriptions')
ax.get_figure().savefig('plots/top25-crimelocs.png')
```

```python
crimes.loc[crimes['Location Description'] == 'LAGOON']
```

```python
traffichist = pd.read_csv('data/traffic_historical.csv', ',', parse_dates = True, index_col='ID', na_values=-1)
traffichist.head()
```

```python
traffichist['SEGMENTID'].value_counts()
```

```python
nonatraf = traffichist.dropna()
seg_counts = nonatraf['SEGMENTID'].value_counts()
ax = seg_counts.plot(kind='bar', figsize=(40, 20))
ax.set_xlabel('Segment')
ax.set_ylabel('Reports')
ax.set_title('Total Reports by Segment')
ax.get_figure().savefig('plots/reports-by-segment.png')
```

```python
weather = pd.read_csv('data/weather.csv', ',', parse_dates = True, index_col='Measurement ID')
weather.head()
```

```python
weather['Measurement Timestamp'] = pd.to_datetime(weather['Measurement Timestamp'])
weather['Station Name'].value_counts()
```

```python
foster = weather.loc[weather['Station Name'] == 'Foster Weather Station'].sort_values('Measurement Timestamp')
foster
```

```python
foster['Air Temperature'].dropna().plot(kind='line', y='Air Temp (Celsius)', x='Date', figsize=(25, 10))
```
